###**Model Training**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Encode categorical features
df_encoded = df.copy()
label_cols = ['Attrition', 'BusinessTravel', 'Department', 'EducationField', 'Gender',
              'JobRole', 'MaritalStatus', 'Over18', 'OverTime']

for col in label_cols:
    df_encoded[col] = LabelEncoder().fit_transform(df_encoded[col])
```

```python
# Define X and y
X = df_encoded.drop(['Attrition', 'EmployeeNumber'], axis=1)
y = df_encoded['Attrition']
```

```python
# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Train-test split (on scaled data)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, stratify=y, test_size=0.2, random_state=42)
```

```python
# Train logistic regression model
model = LogisticRegression(max_iter=2000)  # increased iteration limit
model.fit(X_train, y_train)
```

###**Model Evaluation**

```python
# Make predictions and evaluate
y_pred = model.predict(X_test)
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

**Conclusion**

*   Accuracy of 87%, indicates reliable general performance.
*   High precision (0.89) and recall (0.97) for predicting employees who stay, meaning the model is very effective at identifying non-attrition cases.
*   F1-score of 0.93 for the majority class shows the model handles the dominant group well.


The model successfully learned meaningful patterns from the data using simple preprocessing and logistic regression.
